# Exploratory Analysis on Movies data

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df=pd.read_csv("movies.csv",encoding="ISO-8859-1")
```

```python
#top 5 entries
df.head(5)
```

```python
# Apply describe on df and do the analysis
df.describe()
```

```python
#check missing data
df.isnull().sum()
```

```python
df.info()
```

```python
# Visualization: use factor plot of seaborn for showing the count on 'Day of Week' for size 10
#daywise count we need
plt.figure(figsize=(12,8))
sns.countplot(x="Day of Week",data=df)
plt.xlabel("Day of week",fontsize=15,color="m")
sns.set_style("darkgrid")
plt.ylabel("Count",fontsize=15,color="r")
plt.title("Day of Week VS Count",fontsize=15)
plt.show()
```

 Conclusion: Maximum no of movie are Relasing On friday and after that Wednesday.

```python
df.info()
```

```python
df["Adjusted Gross ($mill)"]=df["Adjusted Gross ($mill)"].str.replace(",","").astype(float)
```

```python
df["Profit ($mill)"]=df["Profit ($mill)"].str.replace(",","").astype(float)
```

```python
# Create below box plot with hue
```

```python
plt.figure(figsize=(20,50))
sns.boxplot(y="Day of Week",x="Profit%",data=df,orient="h",hue="Genre")
plt.show()
```

```python
# Use heatmap to show relationships
```

```python
df.columns
```

```python
# use box plot to check outliers
plt.figure(figsize=(15,50))
sns.boxplot(data=df[['Adjusted Gross ($mill)', 'Budget ($mill)', 'Gross ($mill)', 'Overseas ($mill)', 'Profit ($mill)', 'US ($mill)', 'Gross % US']])
plt.xticks(fontsize=15,rotation=45)
plt.xlabel("Column names",fontsize=15,color="m")
plt.title("Detection of Outlier",fontsize=15)
plt.xlabel("Column names",fontsize=15,color="m")
plt.ylabel("Price range in Million",fontsize=15,color="r")
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=df[['IMDb Rating', 'MovieLens Rating']])
plt.xticks(fontsize=15,rotation=45)
plt.title("Detection of Outlier in Rating",fontsize=15)
plt.xlabel("Column names",fontsize=15,color="m")
plt.ylabel("Ratings",fontsize=15,color="r")
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=df['Runtime (min)'])
plt.xticks(fontsize=15,rotation=45)
plt.title("Detection of Outlier in Runtime",fontsize=15)
plt.xlabel("Runtime",fontsize=15,color="m")
plt.ylabel("Runtime (min)",fontsize=15,color="r")
plt.yticks(fontsize=15)
plt.show()
```

```python
# Calculate per studio profit
df.columns
```

```python
df.groupby("Studio")['Profit ($mill)'].sum().sort_values(ascending=False)
```

```python
# what is average daywise movie length
x=df.groupby("Day of Week")["Runtime (min)"].mean()
```

```python
x
```

```python
plt.figure(figsize=(12,8))
plt.plot(x.index,x.values,marker="o")
plt.xlabel("Day",fontsize=15,color="m")
plt.ylabel("Runtime-min",fontsize=15)
plt.title("Day vs average runtime",fontsize=15)
plt.show()
```

conclusion: Maximum runtime of movie at tuesday,we can see from graph.

```python
#Plot as many as graphs using the features
plt.figure(figsize=(15,8))
sns.countplot(x="Genre",data=df)
plt.xticks(fontsize=15,color="r",rotation=45)
plt.xlabel("GENRE",fontsize=15)
plt.ylabel("COUNT",fontsize=15)
plt.title("Genre Count",fontsize=15)
plt.show()
```

Conclusion:Maximum no of action movie are present in data set.
